```python
!pip install pymupdf > /dev/null

from google.colab import files
import fitz  # PyMuPDF

print("Upload your PDF to reorder pages:")
uploaded = files.upload()
input_pdf = list(uploaded.keys())[0]

doc = fitz.open(input_pdf)
total_pages = len(doc)
doc.close()
print(f"‚úÖ Uploaded '{input_pdf}' with {total_pages} pages (0-indexed).")

order_input = input(f"‚úèÔ∏è Enter new page order (comma-separated, e.g., 2,0,1):\n")
page_order = [int(i.strip()) for i in order_input.split(",")]

doc = fitz.open(input_pdf)
new_doc = fitz.open()
for i in page_order:
    temp = fitz.open()
    temp.insert_pdf(doc, from_page=i, to_page=i)
    new_doc.insert_pdf(temp)
    temp.close()
output_pdf = "reordered_output.pdf"
new_doc.save(output_pdf)
new_doc.close()
doc.close()
print(f"‚úÖ PDF reordered and saved as '{output_pdf}'.")

files.download(output_pdf)
```
